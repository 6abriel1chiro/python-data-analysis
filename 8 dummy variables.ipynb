```python
hidroponia = pd.read_csv('dataset/Hidroponia_original.csv', engine ='python', index_col=0, encoding='latin-1')
```

```python
hidropinia = pd.concat([hidroponia,hortaliza_dummy],axis= 1)
hidropinia = pd.concat([hidroponia,sistema_dummy],axis= 1)
```
